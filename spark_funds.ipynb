## Check point4

```python
master_frame['primary_sector'] = master_frame['category_list'].str.split('|',expand=True,n=1).drop(1,1)
```

```python
master_frame.head()
```

```python
mapping_df = pd.read_excel('./data/mapping.xls')
mapping_df.drop(0,inplace=True)
mapping_df['category_list'] = mapping_df['category_list'].str.replace('0','na')
mapping_df.head()
```

```python
mapping_fl = mapping_df.drop('category_list',1).astype('float')
mapping_fl.index = mapping_df.category_list
mapping_dict = mapping_fl.idxmax(axis=1).to_dict()
```

```python
master_frame['sector_mapping'] = master_frame['primary_sector'].map(mapping_dict)
```

```python
master_frame.head()
```

## Check point 5

```python
funding_min = 5000000
funding_max = 15000000

def country_df(master_frame,country_name):
    df = master_frame[master_frame['country_code'] == country_name]
    df = df[df['funding_round_type']=='private_equity']
    df = df[(df['raised_amount_usd']>funding_min) & (df['raised_amount_usd']<funding_max)]
    return df
```

```python
D1 = country_df(master_frame,country1)
D2 = country_df(master_frame,country2)
D3 = country_df(master_frame,country3)
```

```python
D1.head()
```

### The total number of investments  and amount invested in each main sector 

```python
D1.groupby(['sector_mapping'])['raised_amount_usd'].agg(['count','sum'])
```

```python
D2.groupby(['sector_mapping'])['raised_amount_usd'].agg(['count','sum'])
```

```python
D3.groupby(['sector_mapping'])['raised_amount_usd'].agg(['count','sum'])
```
